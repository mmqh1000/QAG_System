````python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.translate.bleu_score import corpus_bleu

unawareness = [[[] for _ in QA] for _ in range(1,10)]
print(len(unawareness))
for i in range(30):
    # 读取当前结果对应的文件
    with open(path[i], 'r', encoding='utf-8') as f:
        txt = f.read()
    
    # 生成参考集：文档中所有句子的分词列表
    sentences = sent_tokenize(txt)
    references = [[token.lower() for token in word_tokenize(sent) if token.isalpha()] for sent in sentences]
    c=0
    for j in QA:
        # 生成候选集：当前结果所有问题的分词列表
        candidates = []
        try:
            for qa in j[i]:
                tokens = [token.lower() for token in word_tokenize(qa['question']) if token.isalpha()]
                candidates.append(tokens)
        except Exception as e:
            print(QA.index(j),j)
        # 构造BLEU需要的结构（每个问题对应整个文档作为参考）
        list_of_references = [references] * len(candidates)  # 每个候选对应同一组参考
        
        # 参数顺序修正为corpus_bleu(参考集列表, 候选集列表)
        try:
            bleu_score = corpus_bleu(list_of_references, candidates)
        except Exception as e:
            bleu_score=0
        score1 = 1 - bleu_score
        print(score1)
        # 正确获取问题文本并优化正则匹配
        score2 = []
        pattern = r'\b(this|these|those|the above text|as shown in|according to)\b|\d'
        print(c,i)
        for qa in j[i]:
            question = qa['question'].lower()
            # 匹配到禁用词则得0分，否则1分
            m = 0 if re.search(pattern, question) else 1
            score2.append(m)
        print(score2)
        # 计算当前结果的平均得分
        try:
            for n in range(1,10):
                final_score = score1 * n*0.1 + (sum(score2)/len(score2)) * (10-n)*0.1
                unawareness[n-1][c].append(final_score)
        except Exception as e:
            for n in range(1,10):
                final_score=0
                unawareness[n-1][c].append(final_score)
        c=c+1
import pandas as pd
unawareness_s=[[] for _ in range(1,10)]
for i in range(1,10):
    for j in range(len(QA)):
        unawareness_s[i-1].append(sum(unawareness[i-1][j])/len((unawareness[i-1][j])))
data={"model":model_list}
for i in range(1,10):
    data[f"[{i*0.1},{(10-i)*0.1}]"]=unawareness_s[i-1]
data=pd.DataFrame(data)
data.to_csv('unawareness.csv')
````
